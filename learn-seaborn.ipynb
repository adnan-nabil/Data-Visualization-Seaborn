# **Importing necessary libraries and dependencies**
numpy and pandas are imported on previous cell

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **dataset importing**

```python
comp_data = pd.read_csv("/kaggle/input/computersales/ComputerSales.csv")
comp_data.head()
```

# **Distribution Plot**

**Normal way**

```python
sns.distplot(comp_data['Profit'])
```

**Kernel density estimation**

```python
sns.distplot(comp_data['Profit'], kde = False)
```

**Bins**

```python
sns.distplot(comp_data['Sale Price'], bins = 50)
```

# **Joint Plot**

```python
sns.jointplot(x ='Sale Price', y='Profit',data=comp_data)
```

# **some different kind =**

```python
sns.jointplot(x ='Sale Price', y='Profit',data=comp_data, kind = 'scatter')
```

# **use regression**

```python
sns.jointplot(x ='Sale Price', y='Profit',data=comp_data, kind='reg')
```

# ** use kernel density estimation**

```python
sns.jointplot(x ='Sale Price', y='Profit',data=comp_data, kind ='kde')
```

# **Use Hexagon**

```python
sns.jointplot(x ='Profit', y='Age',data=comp_data, kind = 'hex')
```

# **I can save any plot as image using savefig() function of matplotlib**

# **Heatmaps**

```python
comp_data
```

```python
#I need only numbers in dataframe, so craete a new data frame
new_dframe = comp_data[['Age','Profit']]
```

```python
#the new data frame
new_dframe
```

```python
#first i need to have correlaton between columns.
new_dframe_mx = new_dframe.corr()
new_dframe_mx
```

```python
# then use the correlation to see heatmap
#so, for heatmap I need correlation.
sns.heatmap(new_dframe_mx)
```

```python
#little extra 
sns.heatmap(new_dframe_mx, annot=True)
```

```python
new = comp_data[['Profit', 'Sale Price']]
new
```

```python
sns.heatmap(new_mx, annot=True)
```

# heatmap with another way by pivot table
